## Visualization

```python
X=data.drop(['Orders','Gender_F','Gender_M'],axis=1)
```

```python
X.head()
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_std = scaler.fit_transform(X) 
```

```python
X_std=pd.DataFrame(X_std,columns=data.columns[1:36])
X_std.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
n_clusters=[4,5,6,7,8]
for k in n_clusters:
    cluster=KMeans(n_clusters=k,random_state=23)
    predict=cluster.fit_predict(X_std)
    
    score=silhouette_score(X_std,predict,random_state=23)
    print ("For n_clusters = {}, silhouette score is {})".format(k, score))
```

## Model with K = 4

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4,random_state=23) 
km.fit(X_std) 
```

## Clusters

```python
new_data=data.copy(deep=True)
new_data['cluster']=km.labels_
```

```python
new_data.head()
```

```python
np.unique(km.labels_,return_counts=True)    # Size of each cluster
```

```python
sns.countplot(data=new_data,x='cluster')
plt.text(x=-.18,y=2000,s=np.unique(km.labels_,return_counts=True)[1][0])
plt.text(x=.9,y=2000,s=np.unique(km.labels_,return_counts=True)[1][1])
plt.text(x=1.85,y=2000,s=np.unique(km.labels_,return_counts=True)[1][2])
plt.text(x=2.85,y=2000,s=np.unique(km.labels_,return_counts=True)[1][3])
```

##Cluster Centers## 

```python
cc=pd.DataFrame(km.cluster_centers_,columns=new_data.columns[1:36])
cc.head()
```
